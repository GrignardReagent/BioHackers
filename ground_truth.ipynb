```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from collections import Counter
# import tools

image_path = './HackathonData/GroundTruthData/GroundTruth_classification.png'
# Reload the image (ensure it's correct)
image = Image.open(image_path).convert("RGBA")
image_array = np.array(image)

# Get unique colors in the image
unique_colors = np.unique(image_array.reshape(-1, image_array.shape[2]), axis=0)

# Count occurrences of each color in the image
color_counts = Counter(map(tuple, image_array.reshape(-1, image_array.shape[2])))

# Display the most common colors
common_colors = color_counts.most_common(20)  # Show top 20 most frequent colors

# Convert to a readable DataFrame
df_colors = pd.DataFrame(common_colors, columns=["RGBA Color", "Count"])

# Show the color distribution
# tools.display_dataframe_to_user(name="Extracted RGBA Colors", dataframe=df_colors)

# Plot the most common colors for visualization
plt.figure(figsize=(8, 4))
for i, (color, count) in enumerate(common_colors):
    plt.bar(i, count, color=np.array(color) / 255.0, edgecolor='black')
plt.xticks(np.arange(20))
plt.ylabel("Pixel Count")
plt.title("Most Common Colors in the Image")
plt.show()
```
